```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
lending_df = pd.read_csv("loans_full_schema.csv")
lending_df
```

For preprocessing, the first thing we have to do is handle the missing values. That is where we will start.

```python
#identifying and separating null values
lending_df.isnull().sum()
```

```python
#drop the columns from the dataframe that have NaN values
lending_df= lending_df.drop(['disbursement_method','initial_listing_status','num_accounts_30d_past_due','current_accounts_delinq','issue_month','public_record_bankrupt'],axis=1)
lending_df
```

```python
#fill the remaining columns with mean
lending_df= lending_df.fillna(value='Mean')
lending_df
```

Factorizing the data to categorize it

```python
# state
ste_c, ste_v = pd.factorize(lending_df['state'])
lending_df['state'] = ste_c

# homeownership
hmo_c, hmo_v = pd.factorize(lending_df['homeownership'])
lending_df['homeownership'] = hmo_c

# sub_grade
sub_c, sub_v = pd.factorize(lending_df['sub_grade'])
lending_df['sub_grade'] = sub_c

# loan_purpose
lnp_c, lnp_v = pd.factorize(lending_df['loan_purpose'])
lending_df['loan_purpose'] = lnp_c

# grade
gra_c, gra_v = pd.factorize(lending_df['grade'])
lending_df['grade'] = gra_c

# loan_status
lns_c, lns_v = pd.factorize(lending_df['loan_status'])
lending_df['loan_status'] = lns_c

# verified_income
vri_c, vri_v = pd.factorize(lending_df['verified_income'])
lending_df['verified_income'] = vri_c
```

Create a new column called region that uses a lambda and shows which part of the US someone lives in, split up into Pacific, Plains, Midwest, Northeast, and South. Consider having your lambda call a function.  

    a. Is there a better way to do this than with a lambda?

```python
state_codes = lending_df['state'].unique()
state_codes
```

```python
# Defining regions:
northeast = [40, 27, 46, 42, 24, 41, 16, 3, 0, 26, 20,48] 
south = [23, 15, 5, 13, 21, 11, 17, 33, 38, 35, 10, 14, 30, 43, 7, 19]
midwest = [29, 36, 6, 9, 2, 32,18]
pacific = [12, 34, 45, 49, 39, 28, 25, 4, 1,8, 44]
plains = [22,37,31, 47,30]
```

```python
#starting timer
start = time.time()

# lambda approach   
def region(row):
    if row['state'] in northeast:
        return 'northeast'
    if row['state'] in south:
        return 'south'   
    if row['state'] in midwest:
        return 'midwest'
    if row['state'] in pacific:
        return 'pacific'
    if row['state'] in plains:
        return 'plains'
    
lending_df['region'] = lending_df.apply(lambda row: region(row), axis =1)

end = time.time()

print("Time taken for lambda",end-start, 'seconds')

```

```python
# starting timer
start = time.time()

# np.select approach
conditions = [lending_df['state'].isin(northeast),
             lending_df['state'].isin(south),
             lending_df['state'].isin(midwest),
             lending_df['state'].isin(pacific),
             lending_df['state'].isin(plains)]

output = ['northeast','south','midwest','pacific','plains']

lending_df['region'] = np.select(conditions,output)

end = time.time()

print("Time taken to process np.select",end-start, 'seconds')
```

The lambda function is about 14 times slower than the other one.
